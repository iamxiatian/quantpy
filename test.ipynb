```python

def MA(df, n):
    '''
    计算移动均线：Moving Average
    输入为包含Close字段的DataFrame
    '''
    series = pd.Series.rolling(df['Close'], window=n).mean() 
    df['MA_' + str(n)] = series #pd.concat([df, series], axis=1)
    return df
```

```python
MA(df, 5)
df['Open'].plot()
pyplot.show()
```
